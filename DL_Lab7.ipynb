```python
from transformers import pipeline
```

```python
sentiment = pipeline(
    "sentiment-analysis",
    model="textattack/bert-base-uncased-SST-2"
)

texts = ["I love this movie!", "This was terrible."]
print(sentiment(texts))
```

```python
qa = pipeline(
    "question-answering",
    model="deepset/bert-base-cased-squad2"
)

context = "BERT was introduced by Google in 2018."

print(qa(question="Who made BERT?", context=context))
print(qa(question="When was BERT introduced?", context=context))
```
