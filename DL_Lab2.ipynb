```python
import numpy as np
import pandas as pd
```

```python
x=np.array([0.75,0.5,0.80])
w=np.array([0.4,0.3,0.5])
b=-0.850001
lr=0.01
y=1
error=-1
while(error!=0):
  z=0
  for i in range(len(x)):
    z=z+(w[i]*x[i])
  z=z+b
  if(z>=0):
    y_pred=1
  else:
    y_pred=0
  error=y-y_pred
  print("z =",z);
  print("error =",error)
  if(error==0):
    break
  else:
    wn=np.array([])
    for i in range(len(x)):
      wn=np.append(wn,w[i]+(lr*error*x[i]))
    w=wn
    b=b+lr*error
    print("b =",b)
```
